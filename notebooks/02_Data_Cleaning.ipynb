```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Customer Churn.csv")
```

```python
df.dtypes
```

```python
df["TotalCharges"].head(10)
```

```python
(df["TotalCharges"] == " ").sum()
```

```python
# Replace blank spaces with NaN

df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)
```

```python
# Convert TotalCharges to numeric

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df[df["TotalCharges"].isnull()]
```

```python
df[df["TotalCharges"].isnull()][
    ["customerID", "tenure", "MonthlyCharges", "TotalCharges"]
]
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

# Key Findings

1. The `TotalCharges` column was converted from object to float.

2. Eleven records contained missing values represented as blank strings.

3. Investigation showed that all affected customers had zero tenure, indicating they had recently joined and had not yet accumulated charges.

4. Since the missing records represented only 0.16% of the dataset, they were removed to avoid introducing bias.

5. All missing values were successfully eliminated.

6. The cleaned dataset contained 7,032 records with no missing values.

# Conclusion

The dataset was successfully cleaned by correcting data types and removing incomplete records. The resulting dataset is complete, consistent, and suitable for feature engineering and machine learning.
