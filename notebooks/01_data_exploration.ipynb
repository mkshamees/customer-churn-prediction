```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="Churn",
    hue="Churn",
    palette=["#2E8B57", "#DC143C"],
    legend=False
)

plt.title("Customer Churn Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Churn Status")
plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="gender",
    hue="Churn",
    palette="Set2"
)

plt.title("Customer Churn by Gender", fontsize=14, fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Contract",
    hue="Churn",
    palette="Set1"
)

plt.title("Churn by Contract Type", fontsize=14, fontweight="bold")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="InternetService",
    hue="Churn",
    palette="Set3"
)

plt.title("Churn by Internet Service", fontsize=14, fontweight="bold")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="MonthlyCharges",
    bins=30,
    kde=True,
    color="royalblue"
)

plt.title("Distribution of Monthly Charges", fontsize=14, fontweight="bold")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")

plt.show()
```

```python
# Calculate churn percentage by contract type

contract_churn = (
    df.groupby("Contract")["Churn"]
      .value_counts(normalize=True)
      .rename("Percentage")
      .mul(100)
      .reset_index()
)

contract_churn
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=contract_churn[contract_churn["Churn"]=="Yes"],
    x="Contract",
    y="Percentage",
    hue="Contract",
    palette="viridis",
    legend=False
)

plt.title("Churn Rate by Contract Type", fontsize=14, fontweight="bold")
plt.ylabel("Churn Percentage (%)")
plt.xlabel("Contract Type")

plt.show()
```

# Key Findings

1. The dataset originally contained 7,043 customer records with 21 features.

2. There were no missing values reported initially; however, the `TotalCharges` column was stored as an object because it contained blank strings.

3. The target variable (`Churn`) was imbalanced:
   - No Churn: 73.46%
   - Churn: 26.54%

4. Contract type showed a strong relationship with churn:
   - Month-to-month customers had the highest churn rate.
   - Two-year contract customers had the lowest churn rate.

5. Customers with Fiber Optic internet appeared to churn more frequently than DSL users.

6. Customers with shorter tenure showed a much higher likelihood of leaving.

7. Monthly charges appeared to have a positive relationship with churn, suggesting that customers paying higher monthly fees are more likely to leave.

# Conclusion

The exploratory analysis identified several factors associated with customer churn. Contract type, tenure, internet service, and monthly charges appear to be strong predictors of churn. These findings provide a solid foundation for the data cleaning and feature engineering stages that follow.
